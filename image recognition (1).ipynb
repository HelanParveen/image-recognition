```python
import pandas as pd

col_names = "image_name face_id_in_image face_box_top face_box_left face_box_right face_box_bottom face_box_confidence expression_label".split()
df_info.columns = col_names
df_sel = df_info[df_info.face_box_confidence > 45]

df_sel.head()
```

```python
import os
import cv2
a=[]
b=[]
for x, row in df_sel.sample(25000).iterrows():
   img_name=row["image_name"]
   a1=row["face_box_left"]
   a2=row["face_box_right"]
   b1=row["face_box_top"]
   b2=row["face_box_bottom"]
   label=row["expression_label"]
   img_path=os.path.join(images_file_name,img_name)

   image = cv2.imread(img_path)

   if image is not None:
      cropped=image[b1:b2, a1:a2]
      face = cv2.resize(cropped, (65, 65))
      if face is not None:
           a.append(face)
           b.append(label)



```

```python
import numpy as np
from keras.utils import to_categorical
onehot_encoded_b = to_categorical(b)
print(onehot_encoded_b) # Display the one-hot encoded representation
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder



a_train, a_temp, b_train, b_temp = train_test_split(normalized_a, onehot_encoded_b, test_size=0.3, random_state=42)
a_val, a_test, b_val, b_test = train_test_split(a_temp, b_temp, test_size=0.5, random_state=42)

#a_train, a_temp, b_train, b_temp = train_test_split(normalized_a, onehot_encoded_b, size=0.4, random_state=54)
#a_value,a_test, b_value, b_test =train_test_split(a_temp, b_temp, test_size=0.6,random_state=54)
print("training set size:",len(a_train))
print("testing set size:",len(a_test))
print("validation set size:",len(a_val))
```

```python

import numpy as np
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

datagen.fit(a_train)

#datagen.fit(a_train) # fit the ImageDataGenerator on the training data
augumented_data_generator= datagen.flow(a_train, b_train, batch_size=32)

```

```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D,Dense, Flatten
model= Sequential() # create Convolutional Neural Network Model
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(65, 65, 3))) # add 1st convolutional layer
model.add(MaxPooling2D(pool_size=(3, 3))) # add 1st maxpooling layer
model.add(Conv2D(65,(3,3),activation='relu'))# add 1st convolutional layer
model.add(MaxPooling2D(pool_size=(3,3))) # add 2nd maxpooling layer
model.add(Flatten())
model.add(Dense(129, activation= 'relu')) # add fully connected layer
num_of_classes = 7
model.add(Dense(num_of_classes,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy']) # compile the model
history = model.fit(a_train, b_train, epochs=100, batch_size=32, validation_data=(a_val, b_val))
# train the model
```
